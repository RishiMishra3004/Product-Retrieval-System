## In this notebook we are going to do EDA (Exploratory Data analysis)

##  1. Import Tools

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## 2. Load the data

```python
data = pd.read_csv("../data/flipkart_com-ecommerce_sample.csv")
data.T
```

## 3. Basic Statistics

```python
data.head()
```

```python
data.info()
```

## 3. Missing Values and Duplicates

```python
# Check for missing values
print(data.isnull().sum())

# Check for duplicate rows
print(f"Number of duplicate rows: {data.duplicated().sum()}")

# Drop duplicate rows if needed
data = data.drop_duplicates()

# Convert 'No rating available' to NaN
data['product_rating'] = pd.to_numeric(data['product_rating'], errors='coerce')
data['overall_rating'] = pd.to_numeric(data['overall_rating'], errors='coerce')

```

## 4. Univariate Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Distribution of numerical features
data[['retail_price', 'discounted_price', 'product_rating', 'overall_rating']].hist(bins=30, figsize=(12, 8))
plt.show()

```

```python
# Replace '>>', '[', ']' with spaces and strip white spaces
data['cleaned_category_tree'] = (data['product_category_tree']
                                 .str.replace('>>', ' ', regex=False)
                                 .str.replace('[', '', regex=False)
                                 .str.replace(']', '', regex=False)
                                 .str.replace('"', '', regex=False)
                                 .str.strip())
# Remove entries that are now empty or consist only of whitespace
data = data[data['cleaned_category_tree'] != '']

# Split categories by space and explode the lists into individual rows
categories_expanded = data['cleaned_category_tree'].str.split(' ', expand=True).stack()

# Create a dictionary with counts of each cleaned category
category_counts = categories_expanded.value_counts().to_dict()

# Convert the dictionary to a DataFrame for plotting
category_counts_df = pd.DataFrame(list(category_counts.items()), columns=['Category', 'Count'])

# Get the top 20 categories
top_categories = category_counts_df.nlargest(30, 'Count')
top_categories = top_categories[2:]

# Plot the top 20 categories
plt.figure(figsize=(12, 8))
sns.barplot(y='Category', x='Count', data=top_categories, palette='viridis')
plt.title('Top 20 Product Categories by Frequency')
plt.xlabel('Count')
plt.ylabel('Category')
plt.show();
```

```python
# Count occurrences of each category
category_counts = data['product_category_tree'].value_counts()

# Get the top 20 categories
top_categories = category_counts.head(20)

# Filter the dataset to include only top 20 categories
top_categories_data = data[data['product_category_tree'].isin(top_categories.index)]

# Plot the top 20 categories
plt.figure(figsize=(12, 12))
sns.countplot(y='product_category_tree', data=top_categories_data, order=top_categories.index)
plt.title('Top 20 Product Categories by Frequency')
plt.xlabel('Count')
plt.ylabel('Product Category Tree')
plt.show()
```

# 5. Bivariate Analysis

```python
# Scatter plot of retail_price vs discounted_price
plt.figure(figsize=(10, 6))
sns.scatterplot(x='retail_price', y='discounted_price', data=data)
plt.xlabel('Retail Price')
plt.ylabel('Discounted Price')
plt.title('Retail Price vs Discounted Price')
plt.show()

# Heatmap of correlations
plt.figure(figsize=(10, 6))
correlation_matrix = data[['retail_price', 'discounted_price', 'product_rating', 'overall_rating']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Product rating distribution
plt.figure(figsize=(10, 6))
sns.histplot(data['product_rating'].dropna(), bins=30, kde=True)
plt.xlabel('Product Rating')
plt.title('Distribution of Product Ratings')
plt.show()

# Overall rating distribution
plt.figure(figsize=(10, 6))
sns.histplot(data['overall_rating'].dropna(), bins=30, kde=True)
plt.xlabel('Overall Rating')
plt.title('Distribution of Overall Ratings')
plt.show()

# Top brands by frequency
plt.figure(figsize=(12, 6))
sns.countplot(y='brand', data=data, order=data['brand'].value_counts()[:20].index)
plt.title('Top Brands by Frequency')
plt.show()
```

```python
# Analyze description length
data['description_length'] = data['description'].apply(lambda x: len(str(x)))

import matplotlib.pyplot as plt
import seaborn as sns

# Distribution of description lengths
plt.figure(figsize=(12, 6))
sns.histplot(data['description_length'].dropna(), bins=30, kde=True)
plt.title('Distribution of Product Description Lengths')
plt.xlabel('Length of Description')
plt.ylabel('Frequency')
plt.show()
```

```python
# Example of common words in product names
from collections import Counter
import re

def get_most_common_words(text_series):
    words = ' '.join(text_series.dropna().astype(str)).lower()
    words = re.findall(r'\b\w+\b', words)
    return Counter(words).most_common(10)

print("Most common words in product names: \n")
print(get_most_common_words(data['product_name']))
```

```python
# Length of product names
data['name_length'] = data['product_name'].str.len()
plt.figure(figsize=(10, 6))
sns.histplot(data['name_length'], bins=30, kde=True)
plt.xlabel('Product Name Length')
plt.title('Distribution of Product Name Lengths')
plt.show()

# Length of product descriptions
data['description_length'] = data['description'].str.len()
plt.figure(figsize=(10, 6))
sns.histplot(data['description_length'], bins=30, kde=True)
plt.xlabel('Description Length')
plt.title('Distribution of Product Description Lengths')
plt.show()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Initialize CountVectorizer with a limit on the number of features
vectorizer = CountVectorizer(ngram_range=(2, 2), max_features=10000)  # Limit to top 10,000 bigrams
X = vectorizer.fit_transform(data['description'].dropna())

# Extract n-grams and their counts
ngrams = vectorizer.get_feature_names_out()
count_values = X.sum(axis=0).A1  # Sum along the rows and convert to 1D array

# Prepare data for plotting
ngrams_count = pd.DataFrame({'ngram': ngrams, 'count': count_values})
top_ngrams = ngrams_count.nlargest(30, 'count')

# Plotting top 20 bigrams
# Set the style and color palette
sns.set_theme(style='whitegrid');
custom_palette = sns.color_palette("husl", n_colors=len(top_ngrams));

plt.figure(figsize=(12, 12))
sns.barplot(data=top_ngrams, x='count', y='ngram', palette=custom_palette);
plt.title('Top 30 Bigrams in Product Descriptions')
plt.xlabel('Count')
plt.ylabel('Bigram')
plt.show();
```
