## Preprocessing the data['description']

```python
# Convert non-string values to empty strings
data['description'] = data['description'].astype(str)

# Check for any remaining non-string values
print(data['description'].apply(type).value_counts())
```

```python
# Define a set of stop words
stop_words = set([
    'a', 'an', 'the', 'is', 'in', 'on', 'and', 'or', 'to', 'for', 'of', 'with', 'by', 'from', 'at', 'as', 'it', 'that', 'which', 'are', 'was', 'were'
])
```

```python
import re
def preprocess_text(text, stop_words):
    if isinstance(text, str):
        text = text.lower()
        text = re.sub(r'[^\w\s]', '', text)  # Remove punctuation
        text = text.replace('>>', ' ')
        words = text.split()
        words = [word for word in words if word not in stop_words]  # Remove stop words
        text = ' '.join(words)
    else:
        text = ''  # Handle non-string values
    return text
```

```python
# Fill missing values with an empty string or some placeholder
data['description'] = data['description'].fillna('')

# Apply preprocessing with stop word removal
data['processed_description'] = data['description'].apply(lambda x: preprocess_text(x, stop_words))

# Apply preprocessing
data['processed_category'] = data['product_category_tree'].apply(lambda x: preprocess_text(x, stop_words))
```

```python
def build_index(descriptions, categories):
    index = defaultdict(list)
    for i, (desc, cat) in enumerate(zip(descriptions, categories)):
        if isinstance(desc, str):
            words = desc.split()
            for word in words:
                index[word].append(i)
        if isinstance(cat, str):
            words = cat.split()
            for word in words:
                index[word].append(i)
    return index

# Build the index
index = build_index(data['processed_description'], data['processed_category'])
```

```python
from collections import defaultdict

def search(query, index, descriptions, categories, stop_words):
    query_words = preprocess_text(query, stop_words).split()
    if not query_words:
        return []
    
    # Collect all product indices that match the query words
    matched_indices = set()
    for word in query_words:
        if word in index:
            matched_indices.update(index[word])
    
    # Prepare scores for ranking
    scores = defaultdict(int)
    for idx in matched_indices:
        product_description = descriptions[idx]
        product_category = categories[idx]
        
        # Count the occurrences of query words in description and category
        desc_word_count = sum(product_description.count(word) for word in query_words)
        cat_word_count = sum(product_category.count(word) for word in query_words)
        
        # Prioritize category matches by giving it more weight
        scores[idx] += cat_word_count * 10  # High weight for category matches
        scores[idx] += desc_word_count  # Lower weight for description matches
        
        # Check if the product description contains all query words
        if all(word in product_description for word in query_words):
            # Assign a higher score to products containing all words in description
            scores[idx] += 500  # High weight for containing all words in description
    
    # Sort products by score
    ranked_indices = sorted(scores, key=scores.get, reverse=True)
    return ranked_indices
```

```python
import pandas as pd

# Function to retrieve and display top 10 product details
def display_top_results(indices, data, top_n=10):
    # Retrieve rows from data corresponding to the indices
    results_df = data.loc[indices]
    
    # Select relevant columns for display
    display_columns = ['product_name', 'product_url', 'retail_price', 'discounted_price']
    
    # Filter out only relevant columns for better readability
    results_df = results_df[display_columns]
    results_df.reset_index(drop=True, inplace=True)  # Reset index to sequential
    
    # Limit the results to top_n rows
    return results_df.head(top_n)

# Example search query
query = "shampoo for dog"
results = search(query, index, data['processed_description'], data['processed_category'], stop_words)

# Display the top 10 results in a DataFrame
top_results_df = display_top_results(results, data, top_n=5)
top_results_df
```

## Evaluation 

```python
from sklearn.metrics import precision_score, recall_score
import numpy as np

# Assuming 'results' contains the top indices and 'ground_truth' contains the relevant product indices for evaluation
def precision_at_k(results, ground_truth, k=10):
    relevant = set(results[:k])
    return len(relevant & set(ground_truth)) / k

def recall_at_k(results, ground_truth, k=10):
    relevant = set(results[:k])
    return len(relevant & set(ground_truth)) / len(ground_truth)

def mean_average_precision(results_list, ground_truth_list, k=10):
    aps = []
    for results, ground_truth in zip(results_list, ground_truth_list):
        ap = np.mean([precision_at_k(results, ground_truth, k=i) for i in range(1, k+1)])
        aps.append(ap)
    return np.mean(aps)

def mean_reciprocal_rank(results_list, ground_truth_list):
    rr = []
    for results, ground_truth in zip(results_list, ground_truth_list):
        for rank, result in enumerate(results):
            if result in ground_truth:
                rr.append(1 / (rank + 1))
                break
        else:
            rr.append(0)
    return np.mean(rr)
```

```python
queries = ["shampoo", "running shoes"]

# Retrieve the top K results for each query using your search function
query_results = []
ground_truths = []

for query in queries:
    # Get top results using the heuristic model
    top_indices = search(query, index,data['processed_description'])
    query_results.append(top_indices)
    
    # Define the relevant product indices for this query
    if query == "shampoo for dog":
        ground_truths.append([100, 200, 300])  # Example indices of relevant products
    elif query == "running shoes":
        ground_truths.append([400, 500, 600])  # Example indices of relevant products
```

```python
# Evaluate the results
precision = precision_at_k(query_results[0], ground_truths[0], k=10)
recall = recall_at_k(query_results[0], ground_truths[0], k=10)
map_score = mean_average_precision(query_results, ground_truths, k=10)
mrr = mean_reciprocal_rank(query_results, ground_truths)

print(f"Precision@10 for {queries[0]}: {precision}")
print(f"Recall@10 for {queries[0]}: {recall}")
print(f"MAP: {map_score}")
print(f"MRR: {mrr}")
```
